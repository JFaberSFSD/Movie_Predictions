```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import glob
```

```python
#Let's use glob to grab all of our raw TMDB data csv files

q = "Data/final_tmdb_data_*.csv.gz"
chunked_files = glob.glob(q)

#Check the first 5
chunked_files[:5]
```

```python
#Now we will use a for-loop to load and merge all files

df_list = []
for file in chunked_files:
    temp_df = pd.read_csv(file)
    df_list.append(temp_df)
    
## Concatenating the list of dfs into 1 combined
df_combined = pd.concat(df_list)
df_combined
```

```python
#Now let's re-export our combined dataframe to our original combined csv.gz file

df_combined.to_csv('Data/tmdb_results_combined.csv.gz', compression="gzip", index=False)
```
